#Getting the pictures and derivatives

```python
from skimage.io import imshow, imread
from skimage.color import rgb2gray
from skimage import img_as_uint , img_as_ubyte , img_as_bool
from skimage.io import imsave
from skimage.transform import resize
```

Hyper-Parameters

```python
P=64
```

Reading the Raw picture

```python

raw_pic = imread('/content/drive/MyDrive/Colab Notebooks/content/chiko.jpg')
imshow(raw_pic)
```

```python
print("picture shape is : ",raw_pic.shape)
raw_pic
```

Resize to P*P

```python
pic = resize(raw_pic,(P,P))
imshow(pic)
```

GrayScale

```python
pic_gray= rgb2gray(pic)
pic_gray=img_as_ubyte(pic_gray)
imshow(pic_gray)
```

```python
print(pic_gray.shape)
pic_gray
```

Binary

```python
pic_binary = (pic_gray >127)
imshow(pic_binary)
```

```python
print(pic_binary.shape)
pic_binary
```

# Saving the Results

rgb

```python
imsave('/content/drive/MyDrive/Colab Notebooks/content/rgb.jpg',img_as_ubyte(pic))
```
